### *New Column `Age`*

```python
df['Dt_Customer'].sort_values(ascending=False).head(1)
```

<div class="alert alert-block alert-info">
<b>Note:</b> The latest date of customer enrollment was in 2014, so to calculate age of customer that can relate with this dataset is 2015 - customer year's of birth
</div>

```python
current_year = 2015
df['Age'] = current_year - df['Year_Birth']
```

### *Grouping `Age` Column*:

![The Generation Defined](../Project/Screen%20Shot%202023-08-27%20at%2015.30.50.png)

<div class="alert alert-block alert-info">
<b>Note:</b> The range of Ages will be simplified by categorizing it into generations
</div>

```python
def categorize_age(birth):
    if birth >= 1995:
        return "Gen Z <20"  # Age ranges 20 - 
    elif birth >= 1980:
        return "Gen Y 21-35"
    elif birth >= 1965:
        return "Gen X 36-50"
    elif birth >= 1946:
        return "Boomers 51-69"
    else:
        return "Builders >70"

df['Age_Group'] = df['Year_Birth'].apply(categorize_age)
```

### *Grouping `Education` Column*

```python
df['Education'].unique()
```

<div class="alert alert-block alert-info">
<b>Note:</b> Education elements will be simplified into 3 categories which are School, Undergraduate, and Postgraduate
</div>

```python
df['Education'] = df['Education'].replace({'Graduation' : 'Undergraduate',
                                           'Basic'      : 'School',
                                           '2n Cycle'   : 'Postgraduate',
                                           'PhD'        : 'Postgraduate',
                                           'Master'     : 'Postgraduate'})
```

### *Grouping `Living_Status` Column*

```python
df['Living_Status'].unique()
```

<div class="alert alert-block alert-info">
<b>Note:</b> Living Status elements will be simplified into 2 categories which are Alone and Partner
</div>

```python
df['Living_Status'] = df['Living_Status'].replace({'Single'     : 'Alone',
                                                    'Divorced'  : 'Alone', 
                                                    'Widow'     : 'Alone',
                                                    'Alone'     : 'Alone',
                                                    'Absurd'    : 'Alone',
                                                    'YOLO'      : 'Alone',
                                                    'Together'  : 'Partner',
                                                    'Married'   : 'Partner'})
```

### *New Column `Children`, `Is_Parent` and `Family_Size`*

```python
df['Children'] = df['Kidhome'] + df['Teenhome']
df["Is_Parent"] = np.where(df['Children'] > 0, 'Parent', 'Not Parent')
df['Family_Size'] = (df['Living_Status'].replace({'Alone': 1, 'Partner': 2}) + df['Children'])
```

<div class="alert alert-block alert-info">
<b>Note:</b> Column 'Children' is used to calculate the 'Family Size' of customer while 'Is_Parent' will categorize the customer that is having a children

### *New Column `Total_Spent` and `Total_Purchase`*

```python
df['Total_Spent'] = df['Wine'] + df['Fruit'] + df['Meat'] + df['Fish'] + df['Sweet'] + df['Gold']
df['Total_Purchase'] = df['Web_Purchase'] + df['Catalog_Purchase'] + df['Store_Purchase']
```

### *Formulating Market Strategy by Analysing `Correlation` among Columns*

```python
correlation = df[['Income', 'Total_Spent', 'Total_Purchase', 'Discount_Purchase', 'Store_Purchase', 'Catalog_Purchase', 'Web_Purchase', \
                  'Web_Visit', 'Wine', 'Meat', 'Fruit', 'Fish', 'Sweet', 'Gold', 'Age', 'Family_Size', 'Accept_Cmp']].corr()

plt.figure(figsize=(10,10))
mask = np.triu(np.ones_like(correlation))
sns.heatmap(correlation, 
            mask=mask, linewidths=0.5, annot_kws={'size':10},   #type: ignore
            annot=True, fmt='.1f', cmap='vlag',
            vmin=-1, vmax=1);
```

> Summary that can be drawn to `Formulate Market Strategy` after analysing the `heatmap` above:

1. **Income is a proxy for other aspects**
    - This implies that `Income` can serve as a proxy for other related aspects of customer behavior and purchasing patterns. 
    - For example, higher income might be associated with increased `spending`, various types of `purchases`, and engagement with different `products`.
    - This could also imply that higher-income customers are more likely to explore a wide range of offerings and make larger purchases.
    
    <br>
    
1. **Income, Web Purchase, Web Visit**
    - The contrasting correlations between `Income`, `web visits`, and `web purchases` are noteworthy. 
    - The negative correlation with `web visit` and the contrast correlation to the `web visit` could suggest that customers with higher `incomes` might already know what they want to purchase, leading to fewer visits for browsing.
    - They might visit the website with a specific purchase in mind indicating higher engagement and intention to buy.

    <br>

1. **Comparison of Purchase Method**
    - `Web purchase` being the lowest correlation with Income might suggest that the lower the customer income, they are more likely to use Web Purchase Method
    - Moreover, `total purchase` by `in store` has a slightly higher than other method indicating that customer that shop by in store tend to purchase larger number of product.
    - This insight can guide marketing and sales strategies by focusing on promoting specific purchase method based on customer preferences and convenience.

    <br>

1. **Best Selling Product**
    - `Wine` can be said as the most favorite product bought by customer compares to other 5 product
    - `Wine` also the best selling product among all 3 purchase method. This indicates that wine is very flexible product to be sold at any method
    - Product like `Meat`, `Fruit`, `Fish` are less popular in web purchase which is normal because those product required some condition examination before buying it

    <br>

1. **Family Size Shopping Behavior**
    - The `Heatmap` clearly shows that Family Size has a negatively correlation to most of the other aspects
    - This might show that `Bigger family` tends to spend less in shopping at our Supermarket
    - However, interestingly, Bigger family has high correlation with `Discount Purchase` and `Web Visit`
    - This finding shows that despite being less comsuptive, they are very attracted to deals which explains why they are very often to visit the web, perhaps to check ongoing deals by Supermarket
    - This insight will be helpful to formulate market strategy to expand customer segment which having more family members because `Discount Purchase` is not correlated to any other aspect

### *Best Selling Product*

```python
data = df[['Wine', 'Meat', 'Fruit','Fish', 'Sweet', 'Gold']]
grouped_data = pd.DataFrame(data, index=[0])

total_spending = grouped_data.sum(axis=1).values[0]
percentage_data = (grouped_data / total_spending) * 100

labels = percentage_data.columns
sizes = percentage_data.iloc[0]
explode = (0.1, 0, 0, 0, 0, 0)  

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)

plt.title('Percentage Spending on Different Categories')
plt.axis('equal') 
plt.show()
```

<div class="alert alert-block alert-info">
